```python
df1 = pd.read_csv('../input/Cancer.csv')
df1.head()
```

```python
df1.columns = ['Cancer_id', 'Name', 'Category']
```

```python
df2 = pd.read_csv('../input/Cancer_Occurrence.csv')
df2.head()
```

```python
df3 = pd.read_csv('../input/State.csv')
df3.head()
```

```python
df3.columns = ['State_name', 'Percentage population below poverty',
       'Percentage population insured', 'Population', 'Unnamed: 4',
       'Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9']
```

```python
df3.columns
```

```python
merged_data = df1.merge(df2,on=["Cancer_id"])
merged_data.head()
```

```python
Cancer = merged_data.merge(df3, on=["State_name"])
Cancer.head()
```

```python
Cancer.drop(['Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9'], axis=1, inplace=True)
```
